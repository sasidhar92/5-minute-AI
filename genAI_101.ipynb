To quickly demonstrate these example, I am going to use transformers library from hugging face.


```python
# installing transformers library. You can use this library to call different models.
# The beauty of this library is you dont need to worry about the algorithms.
# you can go through the availalbe options in pipeline function for example - sentiment analysis, text generation and use default models.
# if you want to try out other modeles from the hugging face library, you just need to provide that model name to pipeline function

from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier("I love watching and learning from 5 minute AI videos.")
```

```python
generate_text = pipeline("text-generation", model="openai-community/gpt2", trust_remote_code=True)
generate_text("AI in 5 minutes is")
```

```python
question_answer = pipeline("question-answering")
question_answer(

    question = "What is the capital of Kenya?",
    context = "Nairobi is capital of Kenya. Nairobi is also the largest city"
    )
```
